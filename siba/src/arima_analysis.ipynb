```python
import pandas as pd
import numpy as np

# 1️⃣ df_2016 から df_2024 を縦に結合
dfs = [df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022, df_2023, df_2024]
df = pd.concat(dfs, ignore_index=True)

# 2️⃣ "target" 列を作成（"time"を整数型に変換）
df["target"] = df["time"].astype(int)

# 3️⃣ date列から "年" だけを抽出して新たな "year" 列を作成
df["year"] = pd.to_datetime(df["date"], errors='coerce').dt.year

# 4️⃣ 必要な列を選択して、データを整理
df2 = df[["course_len", "year", "place", "target", "race_class", "weather", "ground_state"]].dropna().astype(int)

# 5️⃣ "weather"が"晴れ(1)"、"小雨(0)"、"曇り(2)" かつ "ground_state"が"良(0)"のデータを抽出
filtered_df = df2.query("weather in [0, 1, 2] and ground_state == 0")

# 6️⃣ ピボットテーブルを作成
pivot_table = filtered_df.pivot_table(
    index="year",  # 年ごとに集計
    columns=["course_len", "race_class", "place"],  # 距離 × レースランク × 場所
    values="target",  # 平均タイム
    aggfunc="mean"
)

#thresh を使って、残すべきデータの最小数を指定
#len(pivot_table)-1 だと、欠損値が1つ以下の列のみを残すという意味
pivot_table = pivot_table.dropna(axis=1, thresh=len(pivot_table)-4)


# 7️⃣ スタイリングして表示（ノートブックで見やすくする）
styled_pivot_table = pivot_table.style.format("{:.2f}").background_gradient(cmap="coolwarm")

```
