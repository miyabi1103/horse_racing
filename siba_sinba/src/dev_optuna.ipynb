```python
import nyaggle
from nyaggle.hyper_parameters import list_hyperparams
import pandas as pd
```

```python
params_list = pd.DataFrame(list_hyperparams()).query("objective == 'regression'")
```

```python
#欠損がすくない＝使われている
nulls = params_list.isnull().sum()
nulls[nulls < 5]
```

```python
params_list[nulls[nulls < 20].index]
```

```python
#各列で一番使われてるパラメータを見る
params_list[nulls[nulls < 5].index].mode().T[0].to_dict()
```

```python
# 各列で最頻値を取得し、出現回数が多い順にTop3を取得
top3_params = params_list[nulls[nulls < 5].index].apply(lambda x: x.astype(str).value_counts().head(3))

# 結果を辞書形式で表示
top3_params.T.to_dict()

```
