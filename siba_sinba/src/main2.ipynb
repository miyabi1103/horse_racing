# 精度評価
- popは人気列にかけた場合の回収率、的中率です
- topnはそのモデルが評価した上からn番目にかけた場合の的中回収です
- 例えばtop4の馬単にかけた場合上から4つの馬をかけた、4*3=12通りの的中率、回収率を計算します
- なので、馬単と馬連などの的中率は同じになりますが、回収率は変わります
